Data Preprocessing

```python
x = df.drop(columns=["name","status"])

y = df["status"]
```

```python
x.shape
```

```python
y.shape
```

```python
df.shape
```

splitting data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_test.shape,x_train.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)

x_test = scaler.transform(x_test)
```

```python
print(x_train)
```

MOdel Trainning

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
pred_trai = model.predict(x_train)

```

```python
accu_train = accuracy_score(y_train ,pred_trai)
```

```python
print(accu_train)
```

```python
predict_test = model.predict(x_test)

accu_test = accuracy_score(y_test ,predict_test)
```

```python
print("accu_test",accu_test)
```
